```python
# 10) Sacar los colores RGB unicos en una imagen (cuales rgb existen ej: 0, 0, 0 - 255,255,255 -> 2 colores)
from scipy import ndimage
from scipy import misc
import matplotlib.pyplot as plt
mapache = misc.face()
mapache_color=np.copy(mapache)
mapache_color[:,:,0] = 0
mapache_color[:,:,1] = 0
plt.title("Mapache azul")
plt.imshow(mapache_color)
```

```python
mapache_color=np.copy(mapache)
mapache_color[:,:,1] = 0
mapache_color[:,:,2] = 0
plt.title("Mapache rojo")
plt.imshow(mapache_color)
```

```python
mapache_color=np.copy(mapache)
mapache_color[:,:,0] = 0
mapache_color[:,:,2] = 0
plt.title("Mapache verde")
plt.imshow(mapache_color)
```
